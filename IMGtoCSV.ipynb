```
import numpy as np
from PIL import Image
from matplotlib.pyplot import imshow
import cv2
import os


# IMG_DIR = '/content/drive/My Drive/Computer Vision Project/Images/test/'
IMG_DIR = 'Images/test/solid/D(label=3)/'
# IMG_DIR = '/content/drive/My Drive/Computer Vision Project/Images/test/random/'

for img in os.listdir(IMG_DIR):
  img_array = cv2.imread(os.path.join(IMG_DIR,img), cv2.IMREAD_GRAYSCALE)
  
  img_pil = Image.fromarray(img_array)

  img_28x28 = np.array(img_pil.resize((28, 28), Image.ANTIALIAS))



  # img_array = (img_28x28.flatten())

  # black white switch
  img_array = np.vectorize(lambda x: 255 - x)(img_28x28.flatten())


  img_array  = img_array.reshape(-1,1).T

  # #show image
  # print("grayscale")
  # imshow(np.asarray(img_pil))
  # #
  # #show image
  # print("resize")
  imshow(np.asarray(img_28x28))
  #

  print(img_array)

  with open('test_D_sw.csv', 'ab') as f:

    np.savetxt(f, img_array, delimiter=",")
  
```
